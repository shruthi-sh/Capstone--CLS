## Converting dataset to CSV file

```python
#converting the CNN/dailymail dataset(Sentence-Summary) which is in english to csv file
import csv 
import os
path = "F://capstone dataset//cnn//test"
dir_list = os.listdir(path)
fields = ['Document','Summary','Kannada_summary'] 
filename = "F://capstone dataset//cnn//test.csv"
x={}
with open(filename, 'w',encoding='utf-8') as csvfile: 
    writer = csv.DictWriter(csvfile, fieldnames = fields) 
    writer.writeheader() 
    j=0
    for i in dir_list:
        j+=1
        #print(j)
        if(len(x)!=2):
            if i.endswith(".sent"):
                f = open("F://capstone dataset//cnn/test//"+i, "r", encoding="utf-8")
                z=f.read()
                x['Document']=z
            if i.endswith(".summ"):
                f = open("F://capstone dataset//cnn//test//"+i, "r", encoding="utf-8")
                z=f.read()
                x['Summary']=z
        else:
            #print(x)
            writer.writerow(x) 
            x={}
            if(len(x)!=2):
                if i.endswith(".sent"):
                    f = open("F://capstone dataset//cnn//test//"+i, "r", encoding="utf-8")
                    z=f.read()
                    x['Document']=z
                if i.endswith(".summ"):
                    f = open("F://capstone dataset//cnn//test//"+i, "r", encoding="utf-8")
                    z=f.read()
                    x['Summary']=z
```
