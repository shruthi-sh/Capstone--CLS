```python
model_name = 'google/pegasus-cnn_dailymail'
torch_device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
from transformers import PegasusForConditionalGeneration, PegasusTokenizer,AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = PegasusForConditionalGeneration.from_pretrained(model_name).to(torch_device)
```

```python
import pandas as pd
data=pd.read_csv('/content/drive/My Drive/test_final.csv')
data.head()
```

```python
def pegasus_summ(src_text):
  batch = tokenizer.prepare_seq2seq_batch(src_text, truncation=True, padding='longest', return_tensors='pt').to(torch_device)
  translated = model.generate(**batch)
  tgt_text = tokenizer.batch_decode(translated, skip_special_tokens=True)
  summ="".join(tgt_text)
return summ
```

```python
psummaries=[]
i=0
for text in data['Document']:
  i+=1
  print(i)
  psummaries.append(pegasus_summ(text))
```

```python
data['PSummary']=psummaries
```

```python
data.to_csv('/content/drive/My Drive/ptest_final.csv')
```
